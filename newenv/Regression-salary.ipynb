```python
# Import des librairies
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression
```

```python
# Import des données
salary_data = pd.read_csv("DATA/SalaryData.csv")
```

```python
# Copie des données
df = salary_data.copy()
```

```python
# Entrainement du modèle de prédiction
X = np.array(df["YearsExperience"]).reshape(-1, 1)
y = np.array(df["Salary"]).reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)
```

```python
# Calcul des performances du modèle
model.score(X, y)
```

```python
# Calcul la courbe de régression
y_reg = model.predict(X)
```

```python
# Calcul de prédictions
X_pred = np.array([1.7, 5.5, 10]).reshape(-1, 1)
y_pred = model.predict(X_pred)
```

```python
# Affichage des données graphiquement
plt.scatter(X, y, label="Données d'entraînement")
plt.scatter(X_pred, y_pred, label="Données prédites")
plt.plot(X, y_reg, "r")
plt.xlabel("Années d'experiences")
plt.ylabel("Salaire")
plt.legend()
plt.show()
```
